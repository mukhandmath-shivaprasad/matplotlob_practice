## Adjusting the Plot: Line Colors and Styles

```python
plt.plot(x, np.sin(x - 0), color='blue') # specify color by name
```

```python
plt.plot(x, np.sin(x - 1), color='g') # short color code (rgbcmyk)
```

```python
plt.plot(x, np.sin(x - 2), color='0.75') # Grayscale between 0 and 1
```

```python
plt.plot(x, np.sin(x - 3), color='#FFDD44') # Hex code (RRGGBB from 00 to FF)
```

```python
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) # RGB tuple, values 0 and 1
```

```python
plt.plot(x, np.sin(x - 5), color='chartreuse'); # all HTML color names supported
```

```python
plt.plot(x, np.sin(x - 0), color='blue') 
plt.plot(x, np.sin(x - 1), color='g') 
plt.plot(x, np.sin(x - 2), color='0.75') 
plt.plot(x, np.sin(x - 3), color='#FFDD44') 
plt.plot(x, np.sin(x - 4), color=(1.0,0.2,0.3)) 
plt.plot(x, np.sin(x - 5), color='chartreuse'); 
```

The first adjustment one might wish to make to a plot is to control the line colors and styles.
To adjust the color, you can use the color keyword, which accepts a string argument representing virtually any imaginable color. 
#### If no color is specified, Matplotlib will automatically cycle through a set of default colors for multiple lines.

 adjust the line style using the linestyle keyword

```python
plt.plot(x, x + 0, linestyle='solid')
plt.plot(x, x + 1, linestyle='dashed')
plt.plot(x, x + 2, linestyle='dashdot')
plt.plot(x, x + 3, linestyle='dotted');
```

```python
plt.plot(x, x + 4, linestyle='-')
plt.plot(x, x + 5, linestyle='--')
plt.plot(x, x + 6, linestyle='-.')
plt.plot(x, x + 7, linestyle=':');
```

#### If one wants it to be crisp

```python
plt.plot(x, x + 0, '-g')  #  solid green
plt.plot(x, x + 1,'--c')  # dashed cyan
plt.plot(x, x + 2, '-.k') # dashdot black
plt.plot(x, x + 3, ':r'); # dotted red
```

These single-character color codes reflect the standard abbreviations in the RGB
(Red/Green/Blue) and CMYK (Cyan/Magenta/Yellow/blacK) color systems, commonly used for digital color graphics.

## Adjusting the Plot: Axes Limits

```python
plt.plot(x, np.sin(x))
plt.xlim(-1, 11)
plt.ylim(-1.5, 1.5);
```

```python
plt.plot(x, np.sin(x))

plt.xlim(10,0)
plt.ylim(1.2,-1.2)
```

Reverse the order of arguments to reverse the display

 The **_plt.axis()_** method allows to set the x and y limits with a single call, by passing a list that specifies [xmin, xmax, ymin,ymax]

```python
plt.plot(x, np.sin(x))
plt.axis([-1, 11, -1.5, 1.5])

# note here the potential confusion between axes with an e, and axis with an i
```

The plt.axis() method goes even beyond this, allowing you to do things like auto‐
matically tighten the bounds around the current plot

```python
plt.plot(x, np.sin(x))
plt.axis('tight')
```

```python
plt.plot(x, np.sin(x))
plt.axis('equal')


# ensures an equal aspect ratio so that on your screen, one unit in x is equal to one unit in y
```

## Labeling plots

**labeling of plots: titles, axis labels, and simple legends**

```python
plt.plot(x, np.sin(x))
plt.title("A Sine Curve")
plt.xlabel("x")
plt.ylabel("sin(x)")
```

```python
plt.plot(x, np.sin(x), '-g', label='sin(x)')
plt.plot(x, np.cos(x), ':b', label='cos(x)')
plt.axis('equal')
plt.legend(loc = 0);
```

```python
# help(plt.plot)
```

```python
ax = plt.axes()
ax.plot(x, np.sin(x))
ax.set(xlim = (0,10),
      ylim = (-2,2),
      xlabel = 'x',
      ylabel = 'sin(x)',
       title = 'A simple Plot');
```

```python
ax1 = plt.axes()
x = np.linspace(1,10,100)
ax1.plot(x, np.cos(x), label = 'cos(x)')
ax1.set(xlabel = 'x',
        ylabel = 'cos(x)',
        title = 'Simple Cos(x) plot');
ax1.legend();
```

# Scatter Plots

```python
import numpy as np
import matplotlib as mlp
import matplotlib.pyplot as plt
%matplotlib inline
x = np.linspace(0, 10, 30)
y = np.sin(x)

plt.plot(x, y, 'o', color = 'black');
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5),
             marker, label = "marker ='{0}'".format(marker))
plt.legend()
plt.xlim(0,1.8);
```

```python
rng = np.random.RandomState(0)
for marker in ['o', '.', ',', 'x', '+', 'v', '^', '<', '>', 's', 'd']:
    plt.plot(rng.rand(5), rng.rand(5),
             marker, label = "marker ='{0}'".format(marker))
plt.legend(numpoints =2)
plt.xlim(0,1.8);
```

```python
plt.plot(x, y, '-ok');           # line (-), circle marker (o), black (k)
```

**_Additional keyword arguments to plt.plot specify a wide range of properties of the
lines and markers_**

```python
plt.plot(x, y, '-p', color='gray',
        markersize=15, linewidth=4,
        markerfacecolor='white',
        markeredgecolor='gray',
        markeredgewidth=2)
plt.ylim(-1.2, 1.2);
```

**_A second, more powerful method of creating scatter plots is the plt.scatter function, which can be used very similarly to the plt.plot function_**

```python
plt.scatter(x, y, marker='o');
```
